# **Import Libraries**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#**1. Data Preprocessing**

 **Load Dataset**

I Loaded the Customer Personality Analysis dataset, and perform an initial inspection of the data structure, missing values, and column types.

```python
df = pd.read_csv("data/marketing_campaign.csv", sep="\t")     # Read the dataset

print("Dataset Shape:", df.shape)      # Display the dataset shape (number of rows and columns)
display(df.head())    # Show the first 5 rows of the dataset

print("\nColumn Names:")
print(df.columns) # Print all column names in the dataset

print("\nDataset Info:")
print(df.info()) # Display dataset information such as data types and non-null values

```

```python
print("\n                     Missing Values:")
print(df.isnull().sum())   # Check for missing values in each column


print("\nStatistical Summary:")
display(df.describe()) # Show statistical summary of numerical columns
```

Initial Observations
- The dataset contains customer demographic and   purchasing information.
- Some columns are numerical (e.g., Income, Year_Birth, spending features), while others are categorical (e.g., Education, Marital_Status).
- The "Income" column contains missing values, which will be handled in the preprocessing stage.
- The dataset includes useful features for regression, classification, and clustering tasks

**Handling missing values ​​in Income**

```python
df["Income"] = df["Income"].fillna(df["Income"].median())   # Fill missing values in Income using the median value
```

```python
print(df["Income"].isnull().sum())   # Check missing values after filling Income
```

The "Income" column had 24 missing values, so I replaced them with the median of the column. I used the median because it is less affected by extreme values and is more suitable for income data.

**Feature Engineering**

In this part, I created new variables that summarize important customer information. I calculated "Age" from the birth year, "TotalSpending" from all product spending columns, and "TotalChildren" from the number of children and teenagers at home.

```python
# Create Age from Year_Birth
df["Age"] = 2026 - df["Year_Birth"]

# Create TotalSpending by summing all spending columns
df["TotalSpending"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)

# Create TotalChildren by summing Kidhome and Teenhome
df["TotalChildren"] = df["Kidhome"] + df["Teenhome"]
```

```python
# Check the new features
display(df[["Year_Birth", "Age", "TotalSpending", "TotalChildren"]].head())

# Check TotalSpending calculation
display(df[[
    "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds",
    "TotalSpending"
]].head())

# Check TotalChildren calculation
display(df[["Kidhome", "Teenhome", "TotalChildren"]].head())
```

This plot shows the relationships between Income, Age, and TotalSpending to understand how variables are connected and how customers behave across different features.

This plot shows the distribution of customers income. Most customers have moderate income values, with a few high-income outliers creating a right-skewed distribution.

**Filtering invalid records**

In this step, I removed unrealistic records from the dataset. I filtered out customers with ages below 18 or above 100, and I also removed rows where the income value was equal to zero.

```python
# Count unrealistic ages
print("Rows with age < 18:", (df["Age"] < 18).sum())
print("Rows with age > 100:", (df["Age"] > 100).sum())

# Count zero-income rows
print("Rows with income = 0:", (df["Income"] == 0).sum())
```

```python
# Remove unrealistic ages
df = df[(df["Age"] >= 18) & (df["Age"] <= 100)]

# Remove rows where income is zero
df = df[df["Income"] > 0]
```

```python
# Check if any unrealistic ages or zero-income rows remain after filtering
print("Rows with age < 18:", (df["Age"] < 18).sum())
print("Rows with age > 100:", (df["Age"] > 100).sum())
print("Rows with income = 0:", (df["Income"] == 0).sum())
```

After applying the filtering step, all unrealistic age values and zero-income rows were removed successfully.

**Encoding**

In this step, I converted the categorical variables into numerical form so they can be used in machine learning models. I encoded "Education" as ordinal values based on education level, and I grouped the less frequent categories in "Marital_Status" into "Other" before applying one-hot encoding.

```python
# Encode Education as ordinal values
education_map = {
    "Basic": 0,
    "2n Cycle": 1,
    "Graduation": 2,
    "Master": 3,
    "PhD": 4
}

df["Education"] = df["Education"].map(education_map)
```

```python
# Keep only the 4 most frequent marital status categories
top_4 = df["Marital_Status"].value_counts().nlargest(4).index

df["Marital_Status"] = df["Marital_Status"].apply(
    lambda x: x if x in top_4 else "Other"
)

# Apply one-hot encoding to Marital_Status
df = pd.get_dummies(df, columns=["Marital_Status"], drop_first=True, dtype=int)
```

```python
# Check the dataset after encoding
display(df.head())
print(df.columns)
```

After encoding, the categorical variables were successfully converted into numerical form. The "Education" column was transformed into ordinal values, and "Marital_Status" was converted into multiple binary columns using one-hot encoding. At this point, the dataset is fully prepared for the machine learning tasks.

#**2. Supervised Learning - Regression**

In this section, I used regression models to predict the "TotalSpending" of customers based on their demographic and behavioral features. I prepared the input features and target variable, split the data into training and testing sets, scaled the numerical values, and trained three regression models for comparison.

```python
# Define input features and target variable for regression
X_reg = df.drop(columns=[
    "ID",
    "Dt_Customer",
    "TotalSpending",
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds"
])

y_reg = df["TotalSpending"]

# Check the shape of X and y
print("X shape:", X_reg.shape)
print("y shape:", y_reg.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

# Check the shape after splitting
print("X_train shape:", X_train_reg.shape)
print("X_test shape:", X_test_reg.shape)
print("y_train shape:", y_train_reg.shape)
print("y_test shape:", y_test_reg.shape)
```

Train-test split

I split the dataset into training and testing sets using an 80/20 ratio. The training set contains 1789 samples, while the testing set contains 448 samples. This split allows the models to be trained on one portion of the data and evaluated on unseen data.

```python
from sklearn.preprocessing import StandardScaler

# Scale the regression features using StandardScaler
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

# Check the shape after scaling
print("Scaled X_train shape:", X_train_reg_scaled.shape)
print("Scaled X_test shape:", X_test_reg_scaled.shape)
```

Feature scaling
I applied "StandardScaler" to the regression features to normalize their values before training the models. This step is important because some models, especially linear models, perform better when the input features are on a similar scale.

**Linear Regression**

In this step, I trained a Linear Regression model to predict customer total spending. Then, I evaluated the model using MAE, RMSE, and R².

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Linear Regression

linear_model = LinearRegression()
linear_model.fit(X_train_reg_scaled, y_train_reg)

y_pred_linear = linear_model.predict(X_test_reg_scaled)

mae_linear = mean_absolute_error(y_test_reg, y_pred_linear)
rmse_linear = np.sqrt(mean_squared_error(y_test_reg, y_pred_linear))
r2_linear = r2_score(y_test_reg, y_pred_linear)

print("Linear Regression Results:")
print("MAE:", mae_linear)
print("RMSE:", rmse_linear)
print("R²:", r2_linear)


# Ridge Regression

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_reg_scaled, y_train_reg)

y_pred_ridge = ridge_model.predict(X_test_reg_scaled)

mae_ridge = mean_absolute_error(y_test_reg, y_pred_ridge)
rmse_ridge = np.sqrt(mean_squared_error(y_test_reg, y_pred_ridge))
r2_ridge = r2_score(y_test_reg, y_pred_ridge)

print("\nRidge Regression Results:")
print("MAE:", mae_ridge)
print("RMSE:", rmse_ridge)
print("R²:", r2_ridge)


# Decision Tree Regressor

tree_model = DecisionTreeRegressor(max_depth=5, random_state=42)
tree_model.fit(X_train_reg_scaled, y_train_reg)

y_pred_tree = tree_model.predict(X_test_reg_scaled)

mae_tree = mean_absolute_error(y_test_reg, y_pred_tree)
rmse_tree = np.sqrt(mean_squared_error(y_test_reg, y_pred_tree))
r2_tree = r2_score(y_test_reg, y_pred_tree)

print("\nDecision Tree Regressor Results:")
print("MAE:", mae_tree)
print("RMSE:", rmse_tree)
print("R²:", r2_tree)
```

Regression Results Analysis

Three regression models were trained and evaluated to predict customer total spending.

- Linear Regression achieved an R² score of 0.815, indicating that it explained about 81.5% of the variance in the target variable.
- Ridge Regression produced very similar results to Linear Regression, suggesting that regularization had little effect on model performance.
- Decision Tree Regressor achieved the best performance with an R² score of 0.846 and the lowest prediction errors (MAE and RMSE).

Based on these results, the Decision Tree Regressor was the best-performing model for predicting customer total spending.

# **3. Supervised Learning - Classification**

```python
# Define input features and target variable for classification
X_cls = df.drop(columns=["ID", "Dt_Customer", "Response"])
y_cls = df["Response"]

# Check the shape of X and y
print("X shape:", X_cls.shape)
print("y shape:", y_cls.shape)
```

For the classification task, I used "Response" as the target variable, since it indicates whether a customer accepted the last marketing campaign. I excluded "ID" and "Dt_Customer" because they do not contribute useful predictive information for classification.

```python
# Check class distribution
print("Class distribution:")
print(y_cls.value_counts())

print("\nClass distribution (proportion):")
print(y_cls.value_counts(normalize=True))
```

The class distribution shows that the dataset is imbalanced. Around 85% of customers did not respond to the campaign ("Response = 0"), while only about 15% accepted it ("Response = 1"). This means that accuracy alone is not enough to evaluate the classification models, so additional metrics such as precision, recall, and F1-score are also needed.

```python
# Split the data into training and testing sets
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42, stratify=y_cls
)

# Check the shape after splitting
print("X_train shape:", X_train_cls.shape)
print("X_test shape:", X_test_cls.shape)
print("y_train shape:", y_train_cls.shape)
print("y_test shape:", y_test_cls.shape)
```

I split the dataset into training and testing sets using an 80/20 ratio. Stratified sampling was used to preserve the original class distribution of the target variable in both sets.

```python
# Scale features for Logistic Regression and KNN
scaler_cls = StandardScaler()
X_train_cls_scaled = scaler_cls.fit_transform(X_train_cls)
X_test_cls_scaled = scaler_cls.transform(X_test_cls)

# Check the shape after scaling
print("Scaled X_train shape:", X_train_cls_scaled.shape)
print("Scaled X_test shape:", X_test_cls_scaled.shape)
```

I applied "StandardScaler" to the classification features before training Logistic Regression and K-Nearest Neighbors because these models are sensitive to differences in feature scales. Random Forest does not require scaling, so it will be trained using the original feature values.

**Logistic Regression**

In this step, I trained a Logistic Regression model to predict whether a customer would accept the last marketing campaign (`Response`). Since the dataset is imbalanced, I used `class_weight='balanced'` to give more importance to the minority class during training.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Train Logistic Regression model
log_model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
log_model.fit(X_train_cls_scaled, y_train_cls)

# Predict on test data
y_pred_log = log_model.predict(X_test_cls_scaled)

# Evaluate the model
print("Logistic Regression Results:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_log))
print("Precision:", precision_score(y_test_cls, y_pred_log))
print("Recall:", recall_score(y_test_cls, y_pred_log))
print("F1-score:", f1_score(y_test_cls, y_pred_log))

print("\nClassification Report:")
print(classification_report(y_test_cls, y_pred_log))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_cls, y_pred_log))
```

The Logistic Regression model achieved an accuracy of about **81%**, with a recall of **79%** for the positive class ("Response = 1"). This means the model was able to identify most customers who accepted the campaign, which is important in an imbalanced marketing dataset. However, the precision was lower, indicating that some customers were predicted as responders even though they did not actually respond.

**K-Nearest Neighbors (KNN)**

In this step, I trained a K-Nearest Neighbors classifier to predict whether a customer would respond to the campaign. KNN is a distance-based model, so it uses the scaled classification features.

```python
from sklearn.neighbors import KNeighborsClassifier

# Train KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_cls_scaled, y_train_cls)

# Predict on test data
y_pred_knn = knn_model.predict(X_test_cls_scaled)

# Evaluate the model
print("KNN Results:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_knn))
print("Precision:", precision_score(y_test_cls, y_pred_knn))
print("Recall:", recall_score(y_test_cls, y_pred_knn))
print("F1-score:", f1_score(y_test_cls, y_pred_knn))

print("\nClassification Report:")
print(classification_report(y_test_cls, y_pred_knn))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_cls, y_pred_knn))
```

The KNN model achieved a higher accuracy than Logistic Regression, but its recall for the positive class was much lower. This means that KNN correctly classified many non-responding customers, but it failed to identify a large portion of the customers who actually accepted the campaign. Therefore, despite its higher accuracy, KNN may not be the best model for this imbalanced classification problem.

**Random Forest Classifier**

In this step, I trained a Random Forest classifier to predict whether a customer would respond to the marketing campaign. Random Forest is an ensemble model that combines multiple decision trees and can capture more complex patterns in the data. Unlike Logistic Regression and KNN, it does not require feature scaling.

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_cls, y_train_cls)

# Predict on test data
y_pred_rf = rf_model.predict(X_test_cls)

# Evaluate the model
print("Random Forest Results:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_rf))
print("Precision:", precision_score(y_test_cls, y_pred_rf))
print("Recall:", recall_score(y_test_cls, y_pred_rf))
print("F1-score:", f1_score(y_test_cls, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test_cls, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_cls, y_pred_rf))
```

The Random Forest model achieved the highest accuracy among the three classification models, with an accuracy of about **88%**. It also produced the highest precision for the positive class, which means that when the model predicts that a customer will respond to the campaign, it is often correct. However, the recall remained relatively low, indicating that the model still missed many customers who actually accepted the campaign.

#**4. Unsupervised Learning - Clustering**

I applied K-Means clustering to segment customers into groups based on their spending behavior, income, age, and number of children. The purpose of clustering is to identify hidden customer segments that may help in understanding different customer profiles and improving marketing strategies.

```python
# Select features for clustering
cluster_data = df[["TotalSpending", "Income", "Age", "TotalChildren"]]

# Display the first rows and shape
display(cluster_data.head())
print("Clustering data shape:", cluster_data.shape)
```

For clustering, I selected four customer-related features: "TotalSpending", "Income", "Age", and "TotalChildren". These variables help describe customer purchasing behavior, financial level, age group, and family size, which are useful for customer segmentation.

```python
from sklearn.preprocessing import StandardScaler

# Scale the clustering features
scaler_cluster = StandardScaler()
cluster_scaled = scaler_cluster.fit_transform(cluster_data)

# Check the shape after scaling
print("Scaled clustering data shape:", cluster_scaled.shape)
```

I standardized the clustering features using "StandardScaler" so that all variables contribute equally to the clustering process. This is important because K-Means clustering is based on distances, and variables with larger scales could dominate the results if the data is not standardized.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Apply the Elbow Method to find the optimal number of clusters
inertia_values = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(cluster_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal Number of Clusters")
plt.show()
```

From the elbow plot, I selected 3 clusters because the drop in inertia is much
larger before this point, while after **k = 3** the decrease becomes more
gradual. This suggests that three clusters are enough to capture the main customer patterns without adding unnecessary complexity.

```python
# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(cluster_scaled)

# Show the number of customers in each cluster
print(df["Cluster"].value_counts())

# Preview cluster labels with selected features
display(df[["TotalSpending", "Income", "Age", "TotalChildren", "Cluster"]].head())
```

After choosing the optimal number of clusters, I applied the **K-Means** algorithm with **3 clusters**. A new column called Cluster was added to the dataset to assign each customer to one of the identified customer segments.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce clustering data into 2 dimensions for visualization
pca = PCA(n_components=2)
cluster_pca = pca.fit_transform(cluster_scaled)

# Plot the customer clusters
plt.figure(figsize=(8, 6))
plt.scatter(cluster_pca[:, 0], cluster_pca[:, 1], c=df["Cluster"], cmap="viridis", alpha=0.7)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Customer Segments using K-Means Clustering")
plt.colorbar(label="Cluster")
plt.show()
```

To visualize the clustering results, I used **Principal Component Analysis (PCA)** to reduce the four clustering variables into two components. This makes it easier to display the customer groups in a two-dimensional plot while preserving the main structure of the data.

```python
# Define clustering features
cluster_features = ["TotalSpending", "Income", "Age", "TotalChildren"]

# Compute cluster profile: average values + number of customers in each cluster
cluster_profile = df.groupby("Cluster")[cluster_features].mean().round(2)
cluster_profile["Customer Count"] = df["Cluster"].value_counts().sort_index()

cluster_profile
```

The clustering results identified three customer segments with different characteristics.

- **Cluster 1** is the most valuable group, with the **highest spending and income** and the **lowest number of children**, which suggests strong purchasing power.
- **Cluster 0** represents **older family-oriented customers**, as it has the **highest average age** and **largest number of children**, with moderate spending and income.
- **Cluster 2** includes **lower-income and lower-spending customers**, and it is also the **youngest group** on average.

Overall, the clustering analysis revealed meaningful customer segments that can help support more targeted marketing strategies.
